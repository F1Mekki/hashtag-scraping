<h2>Task 1.1</h2>

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

G = nx.karate_club_graph()

degrees = G.degree() #Dict with Node ID, Degree
nodes = G.nodes()
attributes = nx.get_node_attributes(G, 'club')

print('Nodes:', G.nodes(), '\nEdges:', G.edges(), '\nAttributes:', [(n, {'club': attributes[n]}) for n in nodes])
```

<h2>Task 1.2</h2>

```python
n_color = np.asarray([degrees[n] for n in nodes])
n_size = np.asarray([degrees[n]*50 for n in nodes])

pos = nx.spring_layout(G, seed=6)

# Define a list of colors for the two groups
colors = ['r' if G.nodes[i]['club'] == 'Officer' else 'b' for i in range(len(G.nodes))]

# Draw the graph with node colors
nx.draw(G, with_labels=True, node_color=colors)
plt.show()
```

<h2>Task 1.3</h2>

```python
shortest_path = nx.dijkstra_path(G, 24, 16)
print('Shortest path calculated with dijkstras algorithm:', shortest_path)
```

<h2>Task 1.4</h2>

```python
# Define a dictionary of node colors, where selected nodes are red and others are default
node_colors = {node: 'red' if node in shortest_path else 'blue' for node in G.nodes}

# Draw the graph with the custom node colors
nx.draw(G, with_labels=True, node_color=list(node_colors.values()), node_size=500)
plt.show()
```
